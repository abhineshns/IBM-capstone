## Encoding

```python
#Here all the catogorical variables are nominal we have to apply nominal encoding strategies
df1['ADDRTYPE'].unique()
df3=df1.copy()
```

```python
#Applying one hot encoding on the ADDRTYPE column
df=pd.get_dummies(data=df1,columns=['ADDRTYPE'],drop_first=True)
df.head()
```

```python
len(df['JUNCTIONTYPE'].unique())
```

```python
# Since JUNCTIONTYPE has only 7 catogories we can apply one hot encoding on this variable also
df=pd.get_dummies(data=df,columns=['JUNCTIONTYPE'],drop_first=True)
df.head()
```

```python
len(df['COLLISIONTYPE'].unique())
```

```python
#COLLISIONTYPE is having 10 catogories, If we do one hot encoding here it may lead to high collenearity and high cardinality 
#So choosing mean encoding here
mean=df.groupby(['COLLISIONTYPE'])['SEVERITYCODE'].mean().to_dict()
mean
```

```python

df['COLLISIONTYPE']=df['COLLISIONTYPE'].map(mean)
df.head()
```

```python
len(df['WEATHER'].unique())
```

```python
#Applying mean encoding on Weather attribute
mean=df.groupby(['WEATHER'])['SEVERITYCODE'].mean().to_dict()
mean
```

```python
df['WEATHER']=df['WEATHER'].map(mean)
df.head()
```

```python
print(len(df['ROADCOND'].unique()))
print(len(df['LIGHTCOND'].unique()))
```

```python
# Applying mean encoding on the remaining catogorical attributes
mean=df.groupby(['ROADCOND'])['SEVERITYCODE'].mean().to_dict()
mean
```

```python
df['ROADCOND']=df['ROADCOND'].map(mean)
df.head()
```

```python
mean=df.groupby(['LIGHTCOND'])['SEVERITYCODE'].mean().to_dict()
mean
```

```python
df['LIGHTCOND']=df['LIGHTCOND'].map(mean)
df.head()
```

```python
#Now we have converted all the catogorical columns into numerical columns
```

```python
#saving the encoded data for future use
df.to_csv("Encoded accedent data.csv")
```

# Model creation

# Logistic Regression

```python
import numpy as np
```

```python
#Lets define independent and dependent variables
X = np.asarray(df.drop(columns=['SEVERITYCODE']))
y = np.asarray(df['SEVERITYCODE'])
print(X.shape)
print(y.shape)
```

```python
#Normalizing the dataset

from sklearn import preprocessing
X_norm = preprocessing.StandardScaler().fit(X).transform(X)
X_norm[0:5]
```

```python
#Splitting the data into training and test set 

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X_norm, y, test_size=0.07, random_state=4)
print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape,  y_test.shape)

#We only alloted 7% for the test set because 12803 cases are enough for testing the model 
```

```python
# MODELING LOGISTIC REGRESSION

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
LR = LogisticRegression(C=0.01, solver='liblinear').fit(X_train,y_train)
LR
```

```python
# PREDICTING 

yhat = LR.predict(X_test)
yhat
```

# EVALUATION

```python
# JACCARD INDEX
from sklearn.metrics import jaccard_similarity_score
jaccard_similarity_score(y_test, yhat)
```

```python
from sklearn.metrics import classification_report
print (classification_report(y_test, yhat))
```
